```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import rc

rc('font', **{'family': 'sans-serif', 'sans-serif': ['DejaVu Sans'], 'size': 10})
%matplotlib inline
%config InlineBackend.figure_format = 'retina'  # boost notebook rendering to over 200dpi
```

```python
df_by_cat = pd.read_csv("../data/processed/memory_footprint_by_category.csv")
```

```python
# plots both categories in one plot, no line fit

# limit data size in b. Full data makes points mush together, 
# but trend still holds with lower number of datapoints
max_x = 10000
df_max_filtr = df_by_cat[df_by_cat.disk_size < max_x]

# create empty plot
fig, ax = plt.subplots()
fig.suptitle('Disk vs. DataFrame and ISA objects', fontsize=12, fontweight='bold')

# plot DF data
df_df = df_max_filtr[df_max_filtr.category == 'DataFrame']
# scale to kb
disk_sizes = df_df.disk_size / 1024  
mem_sizes = df_df['size'] / 1024
ax.plot(disk_sizes, mem_sizes, 'o', c='green', alpha=0.25)

# plot ISA object data
df_isa = df_max_filtr[df_max_filtr.category == 'ISA']
disk_sizes = df_isa.disk_size / 1024  
mem_sizes = df_isa['size'] / 1024
ax.plot(disk_sizes, mem_sizes, 'o', c='orange', alpha=0.3)

# set axis labels
ax.set_xlabel('Disk size (kb)')
ax.set_ylabel('Memory size (kb)')

plt.show()
```
